```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
data = [65.2, 67.1, 69.2, 66.9, 62.1, 63.2, 67.2, 76.3, 68.5, 72.1, 82.1, 84.7]
plt.plot(data, marker = "o")
locs, labels = plt.xticks()  # Get the current locations and labels.
plt.xticks(np.arange(0, 1, step=0.2))  # Set label locations.
#plt.xticks(np.arange(3), ['Tom', 'Dick', 'Sue'])  # Set text labels.
plt.xticks([i for i in range(12)], ['Apr \'22', 'May \'22', 'Jun \'22', 'Jul \'22', 'Aug \'22', 'Sep \'22','Oct \'22', 'Nov \'22', 'Dec \'22', 'Jan \'23', 'Feb \'23', 'Mar \'23'],
       rotation=45)  # Set text labels and properties.
plt.xlabel("Month")
plt.ylabel("Score")
plt.title("Your Sustainability Scores in the Past Year")
plt.tight_layout()
plt.savefig("scores.png")
```

```python
data = np.array([65.2, 67.1, 69.2, 66.9, 62.1, 63.2, 67.2, 76.3, 68.5, 72.1, 82.1, 84.7])
data_f = data.mean() + (data-data.mean()) * 0.3 + np.random.randn(data.shape[0]) * 3 - 2
data_a = data.mean() + (data-data.mean()) * 0.1 + np.random.randn(data.shape[0]) * 1 - 7

name = ["Your Score", "Friends Scores", "All Scores"]

plt.plot(data, marker = "o", alpha=0.7, label=name[0])
plt.plot(data_f, marker = "o", alpha=0.7, label=name[1])
plt.plot(data_a, marker = "o", alpha=0.7, label=name[2])
locs, labels = plt.xticks()  # Get the current locations and labels.
plt.xticks(np.arange(0, 1, step=0.2))  # Set label locations.
#plt.xticks(np.arange(3), ['Tom', 'Dick', 'Sue'])  # Set text labels.
plt.xticks([i for i in range(12)], ['Apr \'22', 'May \'22', 'Jun \'22', 'Jul \'22', 'Aug \'22', 'Sep \'22','Oct \'22', 'Nov \'22', 'Dec \'22', 'Jan \'23', 'Feb \'23', 'Mar \'23'],
       rotation=45)  # Set text labels and properties.
plt.xlabel("Month")
plt.ylabel("Score")
plt.legend()
plt.title("Overall Sustainability Scores in the Past Year")
plt.tight_layout()
# plt.savefig("scores.png")
```
